```python
from pathlib import Path
import pandas as pd

# Шлях до папки з CSV-файлами
folder_path = Path("new_parse")

data = pd.DataFrame()

# Перебираємо всі CSV-файли
for csv_file in folder_path.glob("*.csv"):
    print(f"Обробляється файл: {csv_file.name}")
    new_df = pd.read_csv(csv_file)
    data = pd.concat([data, new_df], axis=0)
```

```python
data["recommend"] = np.where(data['recommend'] == 'рекомендує цей товар', 1, 0)
```

```python
data = data[["recommend", "liked", "disliked", "experience", "comment"]]
```

```python
data["experience"].value_counts()
```

```python
def process_exp(word):
    if word == "кілька годин":
        return 1
    elif word == "кілька днів":
        return 2
    elif word == "кілька тижнів":
        return 3
    elif word == "кілька місяців":
        return 4
    elif word == "рік і більше":
        return 5
    else:
        return np.nan

data["experience"] = data["experience"].apply(process_exp)
```

```python
data.head()
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    return str(text).replace("\r", "")

for col in ["liked", "disliked", "comment"]:
    data[col] = data[col].apply(clean_text)
```

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0

def detect_lang(text):
    if len(text) < 3:
        return None
    try:
        lang = detect(text)
        return lang
    except Exception as e:
        return None

df["language"] = df["comment"].apply(detect_lang)
df["language"] = df["language"].apply(lambda x: "uk" if x not in ["uk", "ru"] else ("ru" if x == "ru" else "uk"))
```

```python
import re
from stop_words import get_stop_words
import pymorphy2


# pymorphy2 для обох мов
morph_uk = pymorphy2.MorphAnalyzer(lang='uk')
morph_ru = pymorphy2.MorphAnalyzer()

# Стоп-слова
stopwords_ru = set(get_stop_words("ru"))
stopwords_uk = set(get_stop_words("uk"))
stopwords_ru.discard("не")
stopwords_uk.discard("не")
all_stopwords = stopwords_ru | stopwords_uk

# Очистка
def processing(review):
    review = re.sub('[^a-zA-Zа-яА-ЯіІїЇєЄґҐ]', ' ', review)
    review = review.lower()
    review = review.split()
    review = [word for word in review if word not in all_stopwords]
    return ' '.join(review)

# Лематизація
def lemmatize_uk(text):
    return " ".join([morph_uk.parse(word)[0].normal_form for word in text.split()])

def lemmatize_ru(text):
    return " ".join([morph_ru.parse(word)[0].normal_form for word in text.split()])

# Обробка
for col in ["liked", "disliked", "comment"]:
    df[col] = df[col].apply(processing)
    df[col] = df.apply(
        lambda row: lemmatize_uk(row[col]) if row["language"] == "uk" else lemmatize_ru(row[col]),
        axis=1
    )
    df[col + "_len"] = df[col].apply(lambda x: len(x.split()))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression

tfidf_liked = TfidfVectorizer(max_features=300, ngram_range=(1,2))
tfidf_disliked = TfidfVectorizer(max_features=1000, ngram_range=(1,2))
tfidf_comment = TfidfVectorizer(max_features=500, ngram_range=(1,2))

X_train_tfidf, X_test_tfidf, y_train, y_test = train_test_split(df.drop(["language", "recommend"], axis=1),
                                                                            df["recommend"],
                                                                            test_size=0.2,
                                                                            random_state=42,
                                                                            stratify=df["recommend"])

X_text_liked_tr = pd.DataFrame(tfidf_liked.fit_transform(X_train_tfidf["liked"]).toarray(),
                            columns=["liked_" + f for f in tfidf_liked.get_feature_names_out()])
X_text_disliked_tr = pd.DataFrame(tfidf_disliked.fit_transform(X_train_tfidf["disliked"]).toarray(),
                               columns=["disliked_" + f for f in tfidf_disliked.get_feature_names_out()])
X_text_comment_tr = pd.DataFrame(tfidf_comment.fit_transform(X_train_tfidf["comment"]).toarray(),
                              columns=["comment_" + f for f in tfidf_comment.get_feature_names_out()])


X_text_liked_tst = pd.DataFrame(tfidf_liked.transform(X_test_tfidf["liked"]).toarray(),
                            columns=["liked_" + f for f in tfidf_liked.get_feature_names_out()])
X_text_disliked_tst = pd.DataFrame(tfidf_disliked.transform(X_test_tfidf["disliked"]).toarray(),
                               columns=["disliked_" + f for f in tfidf_disliked.get_feature_names_out()])
X_text_comment_tst = pd.DataFrame(tfidf_comment.transform(X_test_tfidf["comment"]).toarray(),
                              columns=["comment_" + f for f in tfidf_comment.get_feature_names_out()])


x_train = pd.concat([
    X_text_liked_tr.reset_index(drop=True),
    X_text_disliked_tr.reset_index(drop=True),
    X_text_comment_tr.reset_index(drop=True),
    X_train_tfidf[["experience"]].reset_index(drop=True)
], axis=1)

x_test = pd.concat([
    X_text_liked_tst.reset_index(drop=True),
    X_text_disliked_tst.reset_index(drop=True),
    X_text_comment_tst.reset_index(drop=True),
    X_test_tfidf[["experience"]].reset_index(drop=True)
], axis=1)

"""SMOTE USAGE"""
# smote = SMOTE(random_state=42)
# x_train, y_train = smote.fit_resample(x_train, y_train)
# model = XGBClassifier()

"""XGBoostClassifier"""
# neg = np.sum(y_train == 0)
# pos = np.sum(y_train == 1)
# scale = neg / pos
# model = XGBClassifier(scale_pos_weight=scale)

"""LogisticRegression"""
model = LogisticRegression(max_iter=500, class_weight='balanced')

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
y_pred_prob = model.predict_proba(x_test)[:, 1]
y_pred_adj = (y_pred_prob > 0.3).astype(int)
print(classification_report(y_test, y_pred_adj))
```
